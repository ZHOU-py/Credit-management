```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fsvm = pd.read_csv("Fsvm.csv")
LSFsvm = pd.read_csv("LSFsvm.csv")

Fsvm_alpha = pd.read_csv("Fsvm_alpha.csv")
LSFsvm_alpha = pd.read_csv("LSFsvm_alpha.csv")

Fsvm_upSampling = pd.read_csv("Fsvm_upSampling.csv")
LSFsvm_upSampling = pd.read_csv("LSFsvm_upSampling.csv")

Fsvm_lowSampling = pd.read_csv("Fsvm_lowSampling.csv")
LSFsvm_lowSampling = pd.read_csv("LSFsvm_lowSampling.csv")

Fsvm_bagging = pd.read_csv("Fsvm_bagging.csv")
LSFsvm_bagging = pd.read_csv("LSFsvm_bagging.csv")

Fsvm_bagging_upSampling = pd.read_csv("Fsvm_bagging_upSampling.csv")
LSFsvm_bagging_upSampling = pd.read_csv("LSFsvm_bagging_upSampling.csv")

Fsvm_bagging_lowSampling = pd.read_csv("Fsvm_bagging_lowSampling.csv")
LSFsvm_bagging_lowSampling = pd.read_csv("LSFsvm_bagging_lowSampling.csv")

Fsvm_cen_lin = pd.read_csv("../Distance/Fsvm_cen_lin.csv")
LSFsvm_cen_lin = pd.read_csv("../Distance/LSFsvm_cen_lin.csv")

Fsvm_cen_exp = pd.read_csv("../Distance/Fsvm_cen_exp.csv")
LSFsvm_cen_exp = pd.read_csv("../Distance/LSFsvm_cen_exp.csv")

Fsvm_hyp_lin = pd.read_csv("../Distance/Fsvm_hyp_lin.csv")
LSFsvm_hyp_lin = pd.read_csv("../Distance/LSFsvm_hyp_lin.csv")

Fsvm_hyp_exp = pd.read_csv("../Distance/Fsvm_hyp_exp.csv")
LSFsvm_hyp_exp = pd.read_csv("../Distance/LSFsvm_hyp_exp.csv")
```

```python
Fsvm_lin = pd.read_csv("../Kernel/Fsvm_lin.csv")
LSFsvm_lin = pd.read_csv("../Kernel/LSFsvm_lin.csv")

Fsvm_poly = pd.read_csv("../Kernel/Fsvm_poly.csv")
LSFsvm_poly = pd.read_csv("../Kernel/LSFsvm_poly.csv")
```

```python
x = range(1,11,1)
```

```python
x_alpha = range(1,151,1)
```

```python
fig = plt.figure(figsize=(18, 4))

l2=plt.plot(x_alpha,LSFsvm_alpha.iloc[2,:150],'go-',label='LS_FSVM alpha')
l3=plt.plot(x_alpha,Fsvm_alpha.iloc[2,:150],'bo-',label='FSVM alpha')

plt.plot(x_alpha,LSFsvm_alpha.iloc[2,:150],'g--',x_alpha,Fsvm_alpha.iloc[2,:150],'b--')

plt.title('comparation alpha')
plt.xlabel('Index')
plt.ylabel('alpha')
plt.legend()
plt.show()
```

```python
#FSVM和LSFSVM相比，有更多的拉格朗日乘子为零，所以LSFSVM缺乏稀疏性，
```

```python

l2=plt.plot(x,LSFsvm['b_ls'],'go-',label='LS_FSVM b')
l3=plt.plot(x,Fsvm['b_f'],'bo-',label='FSVM b')

plt.plot(x,LSFsvm['b_ls'],'g--',x,Fsvm['b_f'],'b--')

plt.title('comparation b')
plt.xlabel('Index')
plt.ylabel('b')
plt.legend()
plt.show()
```

```python
l2=plt.plot(x,LSFsvm_upSampling['b_ls'],'go-',label='LS_FSVM b')
l3=plt.plot(x,Fsvm_upSampling['b_f'],'bo-',label='FSVM b')

plt.plot(x,LSFsvm_upSampling['b_ls'],'g--',x,Fsvm_upSampling['b_f'],'b--')

plt.title('comparation b')
plt.xlabel('Index')
plt.ylabel('b')
plt.legend()
plt.show()
```
